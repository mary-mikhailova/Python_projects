#### **Задача 1**

Ваша задача написать функцию `get_time_to_conversion`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает среднее время между заходом на сайт и целевым действием.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res`

**Решение**

Напишите свое решение ниже

```python
import json
import csv
from datetime import datetime,time

def date_converter(obj):
    if isinstance(obj, str):
        try:
            return datetime.fromisoformat(obj)
        except ValueError:
            return obj
    elif isinstance(obj, list):
        for i, item in enumerate(obj):
            if isinstance(item, list) and len(item) > 1 and isinstance(item[1], str):
                try:
                    obj[i][1] = datetime.fromisoformat(item[1])
                except ValueError:
                    pass
            else:
                obj[i] = date_converter(item)
    return obj

with open ('sessions.csv', 'r') as file:
  reader= csv.DictReader(file, delimiter= ',')

  session_format=[]

  for row in reader:
      json_str = row['session']
      session = json.loads(json_str, )
      date_converter(session)

      session_dict={'user_id': row['user_id'], 'session': session}

      session_format.append(session_dict)

  def get_time_to_conversion(session_format):
    conversion_times=[]
    for user_session in session_format:
      # получаем сессию из структуры данных
      session=user_session['session']
      entry_time=session[0] # Начало сессии

      # проходим по страницам и ищем "conversion"
      for page in session[2]:
        if page[0]=="conversion":
          conversion_dict={'user_id': user_session['user_id'], 'conversion': page[1]-entry_time}
          conversion_times.append(conversion_dict)

    return round(sum([row['conversion'].total_seconds() for row in conversion_times])/len(conversion_times),2)

res = get_time_to_conversion(session_format)
```

#### **Задача 2**

Ваша задача написать функцию `get_session_duration`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает сколько вообще в среднем длится сессия.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res2`

**Решение**

Напишите свое решение ниже

```python

import json
import csv
from datetime import datetime,time

def date_converter(obj):
    if isinstance(obj, str):
        try:
            return datetime.fromisoformat(obj)
        except ValueError:
            return obj
    elif isinstance(obj, list):
        for i, item in enumerate(obj):
            if isinstance(item, list) and len(item) > 1 and isinstance(item[1], str):
                try:
                    obj[i][1] = datetime.fromisoformat(item[1])
                except ValueError:
                    pass
            else:
                obj[i] = date_converter(item)
    return obj

with open ('sessions.csv', 'r') as file:
  reader= csv.DictReader(file, delimiter= ',')

  session_format=[]

  for row in reader:
      json_str = row['session']
      session = json.loads(json_str, )
      date_converter(session)
      session_dict={'user_id': row['user_id'], 'session': session}
      session_format.append(session_dict)

  def get_session_duration(session_format):
    durations=[]
    for user_session in session_format:
      # получаем сессию из структуры данных
      session=user_session['session']
      #достаем длительность
      duration_dict={'user_id': user_session['user_id'], 'duration':session[1]}
      durations.append(duration_dict)
    return round(sum([duration['duration'] for duration in durations])/len(durations),2)


get_session_duration(session_format)

res2 = get_session_duration(session_format)
```

#### **Задача 3**

Напишите функцию `get_daily_session_count`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает cколько разных заходов в течение дня делает пользователь.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res3`

**Решение**

Напишите свое решение ниже

```python
import json
import csv
from datetime import datetime,time
from collections import defaultdict

def date_converter(obj):
    if isinstance(obj, str):
        try:
            return datetime.fromisoformat(obj)
        except ValueError:
            return obj
    elif isinstance(obj, list):
        for i, item in enumerate(obj):
            if isinstance(item, list) and len(item) > 1 and isinstance(item[1], str):
                try:
                    obj[i][1] = datetime.fromisoformat(item[1])
                except ValueError:
                    pass
            else:
                obj[i] = date_converter(item)
    return obj

with open ('sessions.csv', 'r') as file:
  reader= csv.DictReader(file, delimiter= ',')

  session_format=[]

  for row in reader:
      json_str = row['session']
      session = json.loads(json_str, )
      date_converter(session)
      session_dict={'user_id': row['user_id'], 'session': session}
      session_format.append(session_dict)

  def get_daily_session_count(session_format):
    daily_sessions=[]
    for user_session in session_format:
      # получаем сессию из структуры данных
      session=user_session['session']
      #достаем день
      user_day={'user_id': user_session['user_id'], 'day': session[0].date()}
      daily_sessions.append(user_day)

    cnt_dict=defaultdict(int)
    for session in daily_sessions:
      key=(session['user_id'],session['day'])
      cnt_dict[key] += 1
    return round(sum(cnt_dict.values())/len(cnt_dict),2)

get_daily_session_count(session_format)

res3 = get_daily_session_count(session_format)
```

```python
import json
import csv
from datetime import datetime,time
from collections import defaultdict

def date_converter(obj):
    if isinstance(obj, str):
        try:
            return datetime.fromisoformat(obj)
        except ValueError:
            return obj
    elif isinstance(obj, list):
        for i, item in enumerate(obj):
            if isinstance(item, list) and len(item) > 1 and isinstance(item[1], str):
                try:
                    obj[i][1] = datetime.fromisoformat(item[1])
                except ValueError:
                    pass
            else:
                obj[i] = date_converter(item)
    return obj

with open ('sessions.csv', 'r') as file:
  reader= csv.DictReader(file, delimiter= ',')

  session_format=[]

  for row in reader:
      json_str = row['session']
      session = json.loads(json_str, )
      date_converter(session)
      session_dict={'user_id': row['user_id'], 'session': session}
      session_format.append(session_dict)

  def get_average_session_count(session_format):
    cnt_entry=defaultdict(int)
    for user_session in session_format:
      #считаем заходы
      key=user_session['user_id']
      cnt_entry[key]+=1
    return round(sum(cnt_entry.values())/len(cnt_entry),2)


get_average_session_count(session_format)

res4 = get_average_session_count(session_format)
```
