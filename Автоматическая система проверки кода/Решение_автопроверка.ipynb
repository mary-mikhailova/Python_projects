#### **Задача 1**

Задача не из легких, поэтому начнем по шагам. Для начала напишите функцию `run_tests`, которая прочтет эталонный код и код пользователя из `.py` файлов, выполнит их и сохранит результаты в файл `output.txt`.

**Решение**

Напишите свое решение ниже

```python
from importlib import import_module


test_cases = [(1, 2), (3, 4), (5, 6)]

def run_tests(standard_code, user_code, test_cases, output_file):

  # Чтение кодов из файлов
  standard_module = import_module(standard_code)
  user_module = import_module(user_code)


  test_result = []

  for i, test_case in enumerate(test_cases):
    try:
      student_result=user_module.add_numbers(*test_case)
      etalon_result=standard_module.add(*test_case)

      if student_result == etalon_result:
        test_result.append('Тест пройден')
      else:
        test_result.append('Тест не пройден')
        test_result.append(f'Ожидаемый результат: {etalon_result}')
        test_result.append(f'Результат пользователя: {student_result}')

    except Exception as err:
      with open(output_file, 'w') as f:
        f.write (f'Ошибка при выполнении кода пользователя: {repr(err)}')
      return

  with open(output_file, 'w') as f:
    for line in test_result:
      f.write(line+'\n')



run_tests('etalon', 'user', test_cases, 'output.txt')
```

#### **Задача 3**

Движемся дальше - последнее, что нам нужно проверить - время выполнения. Если время выполнения превышает 5 секунд, уведомите об этом пользователя и остановите выполнение проверки.

Перед написанием кода, выполните следующую ячейку, в ней скачивается новый код пользователя, который проверит работу проверки на время выполнения.

```python
!wget https://gist.github.com/Vs8th/4aba49489a8b0843ef85c4a1a93f923d/raw/user_inf.py
```

**Решение**

Напишите свое решение ниже

```python
from importlib import import_module
import ast
import inspect
import time
import multiprocessing


FORBIDDEN_NAMES = ['eval', 'exec']
output_file = 'output3.txt'
def check_for_malicious_code(code, output_file):
  error_mes = None
  try:
    tree = ast.parse(code)

    for node in ast.walk(tree):
      if isinstance(node, ast.Call) and isinstance(node.func, ast.Name):
        if node.func.id=='eval':
          error_mes = 'Код содержит вызов eval()'
          raise ValueError ('Код содержит вызов eval()')
        elif node.func.id=='exec':
          error_mes = f'Код содержит запрещенный атрибут: {node.func.id}'
          raise ValueError(f'Код содержит запрещенный атрибут: {node.func.id}')
  except SyntaxError as syserr:
    error_mes = f'Код содержит синтаксическую ошибку: {syserr}'
  except ValueError as valerr:
    error_mes = str(valerr)
  except Exception as err:
    error_mes = f'Код содержит вредоносные элементы: {err}'
  return error_mes

def run_tests(standard_code, user_code, test_cases, output_file):

  # Чтение кодов из файлов
  standard_module = import_module(standard_code)
  user_module = import_module(user_code)

  test_result = []

  def timeout(func, args, timeout=5):
    """Запускает функцию с ограничением по времени."""
    def wrapper(queue):
      try:
        result = func(*args)
        queue.put(result)
      except Exception as e:
        queue.put (e)

    queue=multiprocessing.Queue()
    process = multiprocessing.Process(target=wrapper, args=(queue,))
    process.start()
    process.join(timeout=timeout)  # Ждем завершения потока

    if process.is_alive():
        process.terminate()
        raise RuntimeError("Предупреждение: время выполнения превышает 5 секунд.")

    if not queue.empty():
        result = queue.get()
        if isinstance(result, Exception):
            raise result
        return result
    else:
        raise RuntimeError("Процесс завершён, но результат не получен.")


  for i, test_case in enumerate(test_cases):
    try:
      student_result = timeout(user_module.add_numbers, test_case)
      etalon_result = timeout(standard_module.add, test_case)


      if student_result == etalon_result:
        test_result.append('Тест пройден')
      else:
        test_result.append('Тест не пройден')
        test_result.append(f'Ожидаемый результат: {etalon_result}')
        test_result.append(f'Результат пользователя: {student_result}')

    except RuntimeError as rt_err:
        with open(output_file, 'w') as f:
            f.write(str(rt_err))
        return

    except Exception as err:
      with open(output_file, 'w') as f:
        f.write (f'Ошибка при выполнении кода пользователя: {repr(err)}')
      return

  with open(output_file, 'w') as f:
    for line in test_result:
      f.write(line+'\n')

user_module = import_module('user_inf')
user_code = inspect.getsource(user_module)

  # Проверка на вредоносный код

malicious_code_message = check_for_malicious_code(inspect.getsource(user_module), output_file)
if malicious_code_message:
   with open(output_file, 'w') as f:
        f.write("Ошибка при проверке модуля user_malicious:\n")
        f.write(malicious_code_message)
else:
  run_tests('etalon', 'user_inf', test_cases, 'output3.txt')
```
